# Problem 1
Kullanıcıdan bir sayı alınız. Sayı 0'dan büyükse pozitif , 0'dan küçükse negatif sıfıra eşitse sıfır yazdırınız.

```python
# 1.Problemin Çözümünü Buraya Yazınız
sayi1=int(input("Bir sayı giriniz:"))

if sayi1>0:
    print(f"{sayi1} pozitif sayıdır.")
elif sayi1<0:
    print(f"{sayi1} negatif sayıdır.")
else:
    print(f"{sayi1} nötrdür.")
```

 # Problem 2
Kullanıcıdan aldığınız bir sayının mükemmel olup olmadığını bulmaya çalışın.
Bir sayının kendi hariç bölenlerinin toplamı kendine eşitse bu sayıya "mükemmel sayı" denir. Örnek olarak, 6 mükemmel bir sayıdır. (1 + 2 + 3 = 6)


```python
# 2.Problemin Çözümünü Buraya Yazınız
sayi=int(input("Bir sayı giriniz:"))

# Toplam adına bir değişken oluşturup 0'a atayalım

toplam=0

# Girilen sayı ile bir arasında gezelim, bunu range() fonksiyonu ile yaparız. range(başlangı, bitiş, artış miktarı)

for i in range(1,sayi):

    # Eğer sayı i değişkenine tam bölünüyorsa toplama i değerini ekleyelim

    if(sayi%i==0): # Buradaki i değeri range fonksiyonun içindeki sayılari yani i ilk birden başlayıp sayi'ya kadar tanımlanacak.

        toplam +=i

# Sayı değeri toplam değerine eşitse "Mükemmel Sayı", eşit değilse "Mükemmel Sayı Değildir" yazdıralım

if(sayi==toplam):

    print("Mükemmel Sayıdır")

else:

    print("Mükemmel Sayı Değildir")
```

 # Problem 3
Kullanıcıdan aldığınız bir sayının "Armstrong" sayısı olup olmadığını bulmaya çalışın.
Örnek olarak, Bir sayı eğer 4 basamaklı ise ve oluşturan rakamlardan herbirinin 4. kuvvetinin toplamı( 3 basamaklı sayılar için 3.kuvveti ) o sayıya eşitse bu sayıya "Armstrong" sayısı denir.

Örnek olarak : 1634 = 1^4 + 6^4 + 3^4 + 4^4

```python
# 3.Problemin Çözümünü Buraya Yazınız
sayicik= int(input("Bir sayı giriniz:"))

# basamak adına değişken oluşturup sayi değerini string'e çevirelim
basamak=str(sayicik)
toplam = 0

# for döngüsü ile basamak değerlerini gezelim ve rakam adında değişken oluşturup basamağın karakter sayısı ile basamak değerlerini çarpalım
# toplam değerine rakamı ekleyelim
for x in basamak:
    rakam = int(x) ** len(basamak) # len fonksiyonu string değerinin basamak sayısını verir
    toplam += rakam

if toplam ==sayicik:
    print(f"{sayicik} Armstrong Sayı")
else:
    print(f"{sayicik} Armstrong Sayısı Değil")
```

# Problem 5
Markete giren bir YetGen'li sıra beklememek için hızlı kasaya gider. Hızlı kasanın çalışma mantığı şu şekildedir:
- Kullanıcının aldığı her ürünün fiyatını tek tek girmesi gerekir.
- Kullanıcı kasadan çıkmak isterse q tuşuna basmalıdır.
- Kasa girilen fiyatları son fiyat isimli değişkene ekler.
- Değişkeni ekrana bastırır.

İpucu: while döngüsünü sonsuz koşulla başlatın ve kullanıcı q'ya basarsa döngüyü break ile sonlandırın.

```python
# 5.Problemin Çözümünü Buraya Yazınız
toplams=0
while True:
    secim=input("Çıkmak için 'q', diğer durumlarda sayı giriniz: ")

    if secim =="q":
        print("Uygulama Sonlandırıldı")
        break
    else:
        sayi = int(secim)
        toplams+= sayi

print(f"{toplam} TL tutarında tuttu.")
```

# Problem 6

Karşısına çıkan ünlü FizzBuzz sorusuna benzer YetGen sorusunda kendisini yeterince hazır hissetmiyor. Bu soruyu çözmesine destek olmak ister misin?

1’den 100’e kadar sayıları yazdırıyoruz. Ancak eğer sayı 3’e bölünüyor ise Yet, sayı 5’e bölünüyorsa Gen, hem 3’e hem 5’e bölünüyorsa YetGen yazdırıyoruz.

```python
# 7.Problemin Çözümünü Buraya Yazınız
for i in range(1,101):
    
    if i % 3 == 0:
        print("YET")
        
    elif i % 5 ==0 and i % 3 == 0:
        print("YETGEN")

    elif i % 5 ==0:
        print("GEN")
    
    else:
        print(i)
```
